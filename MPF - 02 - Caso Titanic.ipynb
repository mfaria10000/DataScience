```python
#lets treat that with regular expression witch will remove the comma and put first name in front of last name
#I am going to create a funcion to remove the comma and switch first and last names
def fixName(lstLine):
    import re
    regex = re.compile(r'"(.*)(,)(\s.*)"')
    return regex.sub(r'"\3 \1"',lstLine)

```

```python
fixName(file_lines[1])
```

```python
lstDataFixed = fixName(file_lines[1]).split(',')
```

```python
print(len(lst_columns))
print(len(lstDataFixed))
```

```python
print(lst_columns)
print(lstDataFixed)
```

```python
#creating a dictionary with the column names as keys and an empty list as values
dic_file_lines = {}

for column in lst_columns:
    dic_file_lines[column] = []
print(dic_file_lines)
```

```python
#here is the first try at the fixName funcion
for line in file_lines[1:]:
    line_with_fixed_name = fixName(line)
        print(line_with_fixed_name)
```

```python
#here is an exercise for the next step
for line in file_lines[1:]:
    line_with_fixed_name = fixName(line)
    lst_lines_fixed = line_with_fixed_name.split(',')
    print(lst_lines_fixed)
```

```python
#here we create a final list with all passengers name already fixed
lst_final = []
for line in file_lines[1:]:
    line_with_fixed_name = fixName(line)
    lst_lines_fixed = line_with_fixed_name.split(',')
    lst_final.append(lst_lines_fixed)
    
```

```python
print(lst_final[1])
```

```python
print(dic_file_lines)
```

```python
#here we retrieve each attibute for each passenger and append into the dictionary useing the column name to  
#access the dicionary key and the index to access the colum name in the list final fixed passanger
for line in lst_final:
    for index_item,item in enumerate(line):
        column = lst_columns[index_item]
        dic_file_lines[column].append(item)
```

```python
print(dic_file_lines)
```

```python
#here I convert the age items to float and where the age is blank i populate the dictionary with -1
for index_age, age in enumerate(dic_file_lines['Age']):
    if dic_file_lines['Age'][index_age] != '':
        dic_file_lines['Age'][index_age] = float(age)
    else:
        dic_file_lines['Age'][index_age] = -1
```

```python
print(dic_file_lines['Age'])
```

```python
#here i create a list with all the items that have the age diferent than -1 and populating a list with them
lst_age = []
for age in dic_file_lines['Age']:
    if age != -1:
        lst_age.append(age)
```

```python
print(lst_age)
```

```python
#creating a function to calculate the age average from a list
def avg_age(lst_age_parm):
    sum_age = sum(lst_age_parm)
    cnt_age_items = len(lst_age_parm)
    return sum_age /  cnt_age_items
```

```python
#calling the funcion
avg_age(lst_age)
```

```python
avg_age_titanic = avg_age(lst_age)
print(avg_age_titanic)
```

```python
#here i am replacing the -1 age items with the average age
for index_age, age in enumerate(dic_file_lines['Age']):
    if dic_file_lines['Age'][index_age] == -1:
        dic_file_lines['Age'][index_age] = avg_age_titanic
```
