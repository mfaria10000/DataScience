# WORKING WITH SEABORN

```python
import matplotlib as plt
import random as rd
%matplotlib inline
import seaborn as sb
```

```python
#seaborn has several datasets to study, you can get the names as shown below
sb.get_dataset_names()
```

```python
file_titanic = sb.load_dataset('titanic')
file_titanic.head()
```

```python
file_tips = sb.load_dataset('tips')
file_tips.head(10)
```

```python
file_flights = sb.load_dataset('flights')
file_flights.head()
```

```python
file_flights = file_flights.pivot_table(values='passengers',index='month',columns='year')
file_flights
```

# GRAPH TYPES

```python
file_tips.head()
```

```python
#creating a histogram with seaborn
#kde=is the bell curve
sb.distplot(file_tips['tip'],kde=True, bins=30)
```

# CORELATTING GRAPHS

## LINEAR GRAPHS

```python
#joinplot MERGER HISTOGRAM WITH PLOT AND IT IS USEFULL TO COMPARE TWO VARIABLES
#checking the relationship between total bill and tips received
sb.jointplot(x='total_bill',y='tip',data=file_tips)
```

```python
#joinplot MERGER HISTOGRAM WITH PLOT AND IT IS USEFULL TO COMPARE TWO VARIABLES
#chhecking to see if there is a relation between age and fare paid
sb.jointplot(x='age',y='fare',data=file_titanic)
```

```python
#checking to see if there is a relation between table size and tips
sb.jointplot(x='size',y='tip',data=file_tips)
```

# LOOKING AT DATA WITH PAIR PLOT

```python
#useing painplot to check the relatiions on the entire dataset checking the relations between all the variables
sb.pairplot(file_tips)
```

```python
#separating the graphs into categorical and non categorical data
sb.pairplot(file_tips,hue='time')
```

```python
sb.pairplot(file_tips,hue='smoker')
```

```python
sb.pairplot(file_tips,hue='sex',palette='PRGn')
```

# Useing the COUNTPLOT

```python
sb.countplot(x='sex', data=file_titanic)
```

# USEING BOXPLOT

```python
#checking in which day people give more tips
sb.boxplot(x='day',y='tip',data=file_tips)
```

```python
#checking in which day people give more tips by sex
sb.boxplot(x='day',y='tip',data=file_tips,hue='sex')
```

```python
#checking in which day people give more tips by somker
sb.boxplot(x='day',y='tip',data=file_tips,hue='smoker')
```

# USEING THE VIOLIN PLOT

```python
sb.violinplot(x='day',y='tip',data=file_tips)
```

```python
#we can also separate the data by sex
sb.violinplot(x='day',y='tip',data=file_tips,hue='sex')
```

```python
#we can also separate the data by sex and split the violins side by side between the hue value
sb.violinplot(x='day',y='tip',data=file_tips,hue='sex', split=True)
```

# WORKING WITH STRIPLOT

```python
sb.stripplot(x='day',y='tip',data=file_tips)
```

```python
#the jitter parameter spreads the data
sb.stripplot(x='day',y='tip',data=file_tips,jitter=True)
```

# WORKING WITH SWARM PLOT


```python
#this plot does not overlay the pairs
sb.swarmplot(x='day',y='tip',data=file_tips)
```

# OVERLAYING TWO TYPES OF GRAPHICS ON THE SAME CELL

```python
sb.swarmplot(x='day',y='tip',data=file_tips,color='black')
sb.violinplot(x='day',y='tip',data=file_tips)
```

# WORKING WITH CORRELATION

```python
#The lmplot gives a prevision of what the next tips could be,like a linear regression analisys
#here we do the same as above but we separate the hue values in separate graphs to see better the predictions of 
#what the values could be
sb.lmplot(x='total_bill',y='tip',data=file_tips,hue='time',col='time')
```
